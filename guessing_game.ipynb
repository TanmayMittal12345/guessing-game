```python

import random
import time

def guess_game():
    print("\n--- Number Guessing Game ---")
    print("Choose difficulty:")
    print("1. Easy (1-50)")
    print("2. Medium (1-100)")
    print("3. Hard (1-200)")

    while True:
        try:
            choice = int(input("Enter option (1/2/3): "))
            if choice == 1:
                limit = 50
                break
            elif choice == 2:
                limit = 100
                break
            elif choice == 3:
                limit = 200
                break
            else:
                print("Invalid choice. Try again.")
        except:
            print("Enter a valid number.")

    number = random.randint(1, limit)
    attempts = 0
    start = time.time()

    print(f"\nI have chosen a number between 1 and {limit}. Try to guess it!")

    while True:
        try:
            guess = int(input("Your guess: "))
            attempts += 1
        except:
            print("Enter a valid number!")
            continue

        if guess < 1 or guess > limit:
            print(f"Enter a number within range 1-{limit}.")
            continue

        if guess == number:
            end = time.time()
            duration = round(end - start, 2)
            print(f"\nüéâ Correct! The number was {number}.")
            print(f"‚úÖ Attempts: {attempts}")
            print(f"‚è±Ô∏è Time taken: {duration} seconds")
            break
        elif guess < number:
            print("Too low. Try again.")
        else:
            print("Too high. Try again.")

while True:
    guess_game()
    again = input("\nPlay again? (y/n): ").lower()
    if again != "y":
        print("Thanks for playing!")
        break
```
